```python
orders2=orders.loc[orders['CustomerID'].unique()]
```

```python
orders.groupby('CustomerID')['amount_spent'].sum()
```

```python
orders2['Total_amount_spent'].sort_values(ascending=False)
```

```python
orders['Total_amount_spent']=orders.groupby('CustomerID')['amount_spent'].sum()
```

```python
orders2['Group']=pd.qcut(orders2['Total_amount_spent'],
                    q=[0,.75,.95,1], labels=['regular', "preferred", "VIP"])
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
orders2.loc[orders2['Group']=='VIP'].groupby('Country').max()
```

```python
orders2.groupby('Country')['Group'].max()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
orders.loc[orders2[(['Group']=='VIP')&(['Group']=='preferred')]].groupby('Country').max()
```
